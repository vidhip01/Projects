## 2. Resizing the image using Pillow library

```python
from PIL import Image
```

```python
img = Image.open('/content/dog-image.jpg')
img_resized = img.resize((200, 200))
```

Now we have reduced pixes to 200x200 only, we reduced dimensions

```python
print(img_resized.shape)
```

This will show error, first save the image

```python
img_resized.save('dog_image_resized.jpg')
```

```python
# displaying the image from numpy array

img_res = mpimg.imread('/content/dog_image_resized.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

We can see the dimensionas have been reduced

```python
print(img_res.shape)
```

## 3. Converting RGB images to Grayscale image using OpenCV

```python
# importing OpenCV library
import cv2
img = cv2.imread('/content/dog-image.jpg')
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

```python
img.shape
```

Notice now in greyscale image, the three channels are gone, it has only one channel now

cv2.imshow() will display the image. But this will not be allowed in Google Colab.

```python
from google.colab.patches import cv2_imshow
```

```python
# displaying the image
cv2_imshow(grayscale_image)
```

```python
# saving the grayscale image
cv2.imwrite('dog_grayscale_image.jpg', grayscale_image)
```
